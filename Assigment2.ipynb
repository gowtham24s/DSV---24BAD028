```python
#Q1. Rolling Sensor Analysis
import numpy as np
import pandas as pd

def name():
    print("Gowtham S - 24BAD028\n")
name()

#sensor data
sensor = np.array([22.1, 22.5, 23.0, 24.8, 25.1, 26.0, 27.5, 28.0])

s = pd.Series(sensor)

#rolling mean and standard deviation
rolling_mean = s.rolling(window=3).mean()
rolling_std = s.rolling(window=3).std()

#anomaly detection
anomalies = s[abs(s - rolling_mean) > 1.5 * rolling_std]

print("Rolling Mean:\n", rolling_mean)
print("\nRolling Standard Deviation:\n", rolling_std)
print("\nAnomalous Values:\n", anomalies)
```

```python
#Q2. Distance Matrix Computation
name()

points = np.array([[2, 3], [5, 4], [8, 7], [1, 6]])

#euclidean distance matrix using broadcasting
diff = points[:, np.newaxis, :] - points[np.newaxis, :, :]
distance_matrix = np.sqrt(np.sum(diff**2, axis=2))

print("Euclidean Distance Matrix:\n", distance_matrix)
```

```python
#Q3. Monte Carlo Risk Estimation
name()

np.random.seed(10)
returns = np.random.normal(0.001, 0.02, 10000)

#expected return
exp_return = np.mean(returns)

#Probability of loss
prob_loss = np.mean(returns < 0)

lower = np.percentile(returns, 2.5)
upper = np.percentile(returns, 97.5)

print("Expected Return:", exp_return)
print("Probability of Loss:", prob_loss)
print("95% Confidence Interval:", (lower, upper))
```

```python
#Q4. Sparse Interaction Analysis

name()
interactions = np.array([[0,0,5,0],[1,0,0,0],[0,0,0,3],[4,0,0,0]])

#sparsity percentage
sparsity = np.mean(interactions == 0) * 100

#row normalization
row_sum = interactions.sum(axis=1, keepdims=True)
normalized = interactions / row_sum

#top interaction per row
top_interaction = np.max(interactions, axis=1)

print("Sparsity Percentage:", sparsity)
print("\nNormalized Matrix:\n", normalized)
print("\nTop Interaction per Row:", top_interaction)
```

```python
#Q5. Student Performance Analysis
name()
data = {
    "Student": ["A","B","C","D","E"],
    "Math": [78,85,67,90,72],
    "Physics": [74,88,69,92,70],
    "Attendance": [82,90,76,95,80]
}

df = pd.DataFrame(data)

#average score
df["Average"] = df[["Math","Physics","Attendance"]].mean(axis=1)

#grades
df["Grade"] = df["Average"].apply(lambda x: "A" if x>=85 else "B" if x>=70 else "C")

#top performer
top_student = df.loc[df["Average"].idxmax(), "Student"]

print(df)
print("Top Performer:", top_student)
```

```python
#Q6. Time-Series Feature Engineering
name()
timestamps = pd.date_range("2024-01-01", periods=10, freq="6H")
values = [12,18,20,22,19,15,14,17,21,23]

df = pd.DataFrame({"time": timestamps, "value": values})
#extract hour and day
df["hour"] = df["time"].dt.hour
df["day"] = df["time"].dt.day

#cyclic features
df["hour_sin"] = np.sin(2*np.pi*df["hour"]/24)
df["hour_cos"] = np.cos(2*np.pi*df["hour"]/24)

#peak activity hour
peak_hour = df.loc[df["value"].idxmax(), "hour"]

print(df)
print("Peak Hour:", peak_hour)
```

```python
#Q7. Data Leakage Detection
name()
train = pd.DataFrame({
    "id": [1,2,3,4],
    "score": [65,70,80,90],
    "result": [0,0,1,1]
})

test = pd.DataFrame({
    "id": [3,5],
    "score": [80,75],
    "result": [1,0]
})

#overlapping records
overlap = pd.merge(train, test, on="id")

#leakage percentage
leakage_percent = (len(overlap) / len(test)) * 100

#columns causing leakage
leakage_columns = ["result"]

print("Overlapping Records:\n", overlap)
print("Leakage Columns:", leakage_columns)
print("Leakage Percentage:", leakage_percent)
```

```python
#Q8. Memory Optimization
name()
df = pd.DataFrame({
    "A": np.random.randint(0, 1000, 10000),
    "B": np.random.rand(10000),
    "C": ["category"] * 10000
})

before = df.memory_usage(deep=True).sum()

df["A"] = df["A"].astype("int16")
df["B"] = df["B"].astype("float32")
df["C"] = df["C"].astype("category")

after = df.memory_usage(deep=True).sum()

print("Before:", before)
print("After:", after)
```

```python
#Q9. Uncertainty Visualization (Matplotlib)
import matplotlib.pyplot as plt
name()

x = np.arange(1,11)
y = np.array([10,12,15,18,20,22,24,23,25,27])
error = np.array([1,1.2,1.5,1.3,1.4,1.6,1.8,1.7,1.9,2.0])

plt.plot(x, y)
plt.fill_between(x, y-error, y+error, alpha=0.3)

i = np.argmax(error)
plt.scatter(x[i], y[i])
plt.text(x[i], y[i], "Max Uncertainty")

plt.show()
```
